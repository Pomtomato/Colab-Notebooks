```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

from scipy.optimize import minimize
```

```python
# minimize f(x) = (x-3)**2
# traditional way->
# df(x)/dx = 2*(x-3)
# 0 = 2*(x-3)
# x = 3 --> will minimize f(x) = (x-3)**2
def f(x):
  return (x-3)**2

res = minimize(f,2)
res
```

```python
f = lambda x: (x[0] - 1)**2 + (x[1] - 2.5)**2
constraints = ({'type': 'ineq', 'fun': lambda x: x[0]-2*x[1]+2},
               {'type': 'ineq', 'fun': lambda x: x[0]-2*x[1]+6},
               {'type': 'ineq', 'fun': lambda x: x[0]+2*x[1]+2},
               )
bounds = ((0, None), (0, None)) # x >=0, y >=0 or x--> (0,inf), y-->(0,inf)
res = minimize(f, (2,0), bounds=bounds, constraints= constraints) # initial guess->(2,0)
```

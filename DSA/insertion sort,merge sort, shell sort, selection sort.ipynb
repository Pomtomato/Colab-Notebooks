```python
# @title shell sort (optimization of insertion sort)
def shell_sort(arr):
  n = len(arr)
  div = 2
  gap = n // 2

  if n < 3:
    return []

  while gap > 0:
    index = []
    for i in range(gap,n):
      anchor = arr[i]
      j = i
      while j >= gap and arr[j-gap] >= anchor:
        if arr[j-gap] == anchor:
          index.append(j-gap)
        arr[j] = arr[j-gap]
        j -= gap
      arr[j] = anchor

    index = list(set(index))
    #index.sort()
    shell_sort(index)

    if index:
      for i in index[-1::-1]:
        del arr[i]
    div *= 2
    n = len(arr)
    gap = n // div

  # if index:
  #   c = 0
  #   for i in index:
  #     if c == 0 :
  #       del arr[i]
  #       c += 1
  #     else:
  #       i = i - c
  #       c += 1
  #       del arr[i]

  return index

if __name__ == "__main__":
  #elements = [21,38,29,17,4,25,11,32,9]
  #elements = [2,1,5,7,2,0,5,1,2,9,5,8,3]
  elements = [2, 1, 5, 7, 2, 0, 5, 1, 2, 9,  5, 8, 3]

  index = shell_sort(elements)
  print(elements)
```

```python
# @title selection sort
def find_min(arr):
  min = 1000000000000000000
  for i in range(len(arr)):
    if arr[i] < min:
      min = arr[i]
  return min

def selection_sort(arr):
  size = len(arr)
  for i in range(size-1):
    min_idx = i
    #anchor = arr[i]
    for j in range(min_idx+1,size):
      if arr[j] < arr[min_idx]:
        min_idx = j
    if i != min_idx:
      arr[i], arr[min_idx] = arr[min_idx],arr[i]

if __name__ == "__main__":
  elements = [21,38,29,17,4,25,11,32,9]

  selection_sort(elements)
  #print(find_min(elements))
  print(elements)
```

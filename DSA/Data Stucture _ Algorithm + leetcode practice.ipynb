```python
# two sums
nums = [2,4,6,8]
target = 10
hashmap = {}
for i in range(len(nums)):
    complement = target - nums[i]
    if complement in hashmap:
        print(i, hashmap[complement])
    hashmap[nums[i]] = i

print(hashmap)
```

```python
# roman to integer
s = "LVIII" # 58
conversion = {"I":1, "V":5, "X":10, "L":50, "C":100, "D":500, "M":1000}

ans = 0
for i in range(len(s)):
  if i < len(s)-1 and conversion[s[i]] < conversion[s[i]]:
    ans -= conversion[s[i]]
  else:
    ans += conversion[s[i]]

print(ans)
```

```python
# Linked List example
class Node:
  def __init__(self, data=None, next=None):
    self.data = data
    self.next = next

class LinkedList:
  def __init__(self):
    self.head = None

  def insert_at_beginning(self, data):
    node = Node(data, self.head)
    self.head = node

  def print(self):
    if self.head is None:
      print("Linked list empty")
      return

    itr = self.head
    llstr = ""

    while itr:
      llstr += str(itr.data) + '--->'
      itr = itr.next

    print(llstr)

if __name__ == '__main__':
  ll = LinkedList()
  ll.insert_at_beginning(5)
  ll.insert_at_beginning(89)
  ll.print()
```

```python
## Binary search
A = [11,22,32,43,55,56,60,67,70,74,80,81,82,86,89,93,99]
x = 111
def binary_search(A, x):
  left = 0
  right = len(A) - 1
  while (left <= right):
    mid = (left + right) // 2
    if A[mid] == x:
      return f"target is: {A[mid]} at the index: {mid} and is element no {mid+1}"
    if A[mid] < x:
      left = mid + 1
    else:
      right = mid - 1

  return f"Target {x} is not in the array"

print(binary_search(A,x))
```

```python
## selection sort
A = [5,9,45,2,465,98,78,22,35,67,18,75,75] # 13
B = [9,3,1,5]
def selection_sort(A):
  for i in range(len(A)-1):
    index_min = i
    for j in range(i+1, len(A)):
        if A[j] < A[index_min]:
          index_min = j
    if index_min != i:
      temp = A[i]
      A[i] = A[index_min]
      A[index_min] = temp

  return A

def selection_sort2(A):
  for i in range(len(A)-1):
    # tmp = A[i]
    for j in range(i+1, len(A)):
      if A[j] < A[i]:
        tmp = A[i]
        A[i] = A[j]
        A[j] = tmp

  return A


print(selection_sort2(A))


```

```python
# bubble sort
A = [5,9,45,2,465,98,78,22,35,67,18,75,75] # 13
B = [9,3,1,5]
C = [10,5,2,8,7]

def bubble_sort(A):
  for i in range(len(A)):
    for j in range(0, len(A)-1-i): ############
      if A[j] > A[j+1]: ##############
        temp = A[j]
        A[j] = A[j+1]
        A[j+1] = temp

  return A

print(bubble_sort(A))
```

```python
# insertion sort
A = [5,9,45,2,465,98,78,22,35,67,18,75,75] # 13
B = [9,3,1,5]
C = [10,5,2,8,7]
D = [5,4,3,2,1]

def insertion_sort(x):
  for i in range(1,len(x)):
    item = x[i]
    j = i - 1

    while j >= 0 and x[j] > item:
      x[j+1] = x[j]
      j = j - 1

    x[j+1] = item

  return x


print(insertion_sort(A))
```

```python
def fib(n):
  if n == 0:
    return 0
  if n == 1 or n == 2:
    return 1
  else:
    return fib(n-1) + fib(n-2)

n = 5;
print("Fibonacci series of n numbers is :",end=" ")
for i in range(0,10):
  print(fib(i),end=" ")
```

```python
def fact(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * fact(n-1)

def fact_tail(n, a=1):
  if n <= 1:
    return a

  return fact_tail(n-1, n * a) # accumalating the factorial value seperately

print("Factorial series of n numbers is :",end=" ")
print(fact_tail(5))
```

```python
# @title binary search
def linear_search(numbers_list, number_to_find):
    for index, element in enumerate(numbers_list):
        if element == number_to_find:
            return index
    return -1


def binary_search(numbers_list, number_to_find):
    left_index = 0
    right_index = len(numbers_list) - 1
    mid_index = 0

    while left_index <= right_index:
        mid_index = (left_index + right_index) // 2
        mid_number = numbers_list[mid_index]

        if mid_number == number_to_find:
            return mid_index

        if mid_number < number_to_find:
            left_index = mid_index + 1
        else:
            right_index = mid_index - 1

    return -1



def binary_search_recursive(numbers_list, number_to_find, left_index, right_index):
    if right_index < left_index:
        return -1

    mid_index = (left_index + right_index) // 2
    if mid_index >= len(numbers_list) or mid_index < 0:
        return -1

    mid_number = numbers_list[mid_index]

    if mid_number == number_to_find:
        return mid_index

    if mid_number < number_to_find:
        left_index = mid_index + 1
    else:
        right_index = mid_index - 1


    return binary_search_recursive(numbers_list, number_to_find, left_index, right_index)

# exercise
def poppin(numbers_list,number_to_find,left_index,right_index):
    index = binary_search_recursive(numbers_list, number_to_find, left_index, right_index)
    indices = [index]

    i = index - 1
    while i >= 0:
      if numbers_list[i] == number_to_find:
        indices.append(i)
      else:
        break
      i = i - 1

    i = index + 1
    while i<len(numbers_list):
      if numbers_list[i] == number_to_find:
        indices.append(i)
      else:
        break
      i = i + 1

    return sorted(indices)

# def knockin(numbers_list,number_to_find,left_index,right_index):
#   index = binary_search_recursive(numbers_list, number_to_find, left_index, right_index)
#   indices = [index]

#   left = numbers_list[:index]
#   right = numbers_list[index+1:]
#   i = index - 1
#   if number_to_find in left:
#     return knockin(left, number_to_find, 0, len(left))
#   else:
#     return indices


#   elif number_to_find in left:
#      knockin(left,number_to_find,0,len(left))


if __name__ == '__main__':
    numbers_list = [12, 15, 17, 19, 21, 24, 45, 67]
    numbers_list = [1,4,6,9,11,15,15,15,17,21,34,34,56]
    number_to_find = 15

    index = poppin(numbers_list, number_to_find, 0, len(numbers_list))
    print(index)
```

```python
# @title bubble sort+exercise
def bubble_sort(elements, key=""):
  size = len(elements)

  for j in range(size-1):
    swapped = False
    for i in range(size-1-j):

      if elements[i][key] > elements[i+1][key]:
        temp = elements[i][key]
        elements[i][key] = elements[i+1][key]
        elements[i+1][key] = temp
        swapped = True # If any swaps were performed in one pass of the given list; (range(size-1-j))

    if not swapped: # without this O(n^2). If no swaps were done in first pass, list is sorted
      break # so now O(n)

if __name__ == "__main__":
  #elements = [9,5,2,1,67,34,88,34]
  #elements = ['mona', 'dhaval', 'raiyan', 'paval', 'alvin']
  elements = [
        { 'name': 'mona',   'transaction_amount': 1000, 'device': 'iphone-10'},
        { 'name': 'dhaval', 'transaction_amount': 400,  'device': 'google pixel'},
        { 'name': 'kathy',  'transaction_amount': 200,  'device': 'vivo'},
        { 'name': 'aamir',  'transaction_amount': 800,  'device': 'iphone-8'},
    ]

  bubble_sort(elements,key="name")
  print(elements)
```

```python
# @title quick sort (Hoare partition)
# pivot number-> eg. [11,4,2,9,15,20,33,21] -> [4,2,9,11,15,20,33,15,21] here 11 is partitioning number
# pivot number can be-> leftmost/ rightmost/ median element etc
# move pivot number to appropriate place/middle
# make a left and right partition,
# where everyting on the left is less than pivot value and
# everything on right is greater than pivot value
# partioning schemes-> 1. Hoare Partition, 2. Lomuto Partition
# continue this partitioning recursively
def swap(a, b, arr):
  if a != b:
    arr[a], arr[b]= arr[b], arr[a]
    # temp = arr[a]
    # arr[a] = arr[b]
    # arr[b] = temp
  #return arr

def partition(elements,start,end):
  pivot_idx = start # initialy 0
  pivot = elements[pivot_idx]

  # start = pivot_idx + 1   # hard-coded without passing in start and end into the function
  # end = len(elements) - 1

  while start < end:
    while start < len(elements) and elements[start] <= pivot:
      start += 1

    while elements[end] > pivot:
      end -= 1

    if start < end:
      swap(start, end, elements)

    #print(start, end) # ->2 4 ->4 3
    #print(elements) #[11, 9, 2, 7, 29, 15, 28] ->[11, 9, 2, 7, 29, 15, 28]

  swap(end,pivot_idx,elements)

  #return elements
  return end


def quick_sort(elements,start,end):
  if start < end:
    p_idx = partition(elements,start,end)
    quick_sort(elements, start, p_idx-1) # left_partition *here start is always 0, because we didn't return start idx from partition and its using the initially passed in 0
    quick_sort(elements, p_idx+1, end) # right partition *here p_idx is still 3, which we got from original partition
                                      # p_idx isn't changed by (quick_sort)left_partition, because that quick_sort call created its own separate sets of variables

if __name__ == "__main__":
  elements = [11,9,29,7,2,15,28]
  partition(elements, 0, len(elements)-1 )
  print(elements)
  # quick_sort(elements)
  # print(elements)
```

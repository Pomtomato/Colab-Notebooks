```python
# @title __ new __ method
# __new__ method is called internally/implicitely everytime a new object is created
class new_method:
  def __new__(cls):
    print("A new object is created using __new__ magic method, __new__ magic method is called")
    inst = object.__new__(cls)
    return inst

  def __init__(self):
    print("Object initialized using __init__ magic method, __init__ magic method is called")
    self.name = "Raiyan"

obj = new_method()
```

```python
# @title __ str __ method
# __str__ will be called internally when 'str' function is called

class employee:
  def __init__(self):
    self.name = "Raiyan"
    self.salary = 10000

  def __str__(self): # overloading the __str__ magic method
    return 'Name: ' + self.name + ', Salary: $' + str(self.salary)

empl = employee()
print(empl)
```

```python
# @title __ add __ method
class distance:
  def __init__(self, x=None, y=None):
    self.ft = x
    self.inch = y

  def __add__(self, other):
    temp = distance()
    temp.ft = self.ft + other.ft
    temp.inch = self.inch + other.inch

    if temp.inch >= 12:
      temp.ft += 1
      temp.inch -= 12
      return temp

  def __str__(self):
    return 'ft: ' + str(self.ft) + ', inch: ' + str(self.inch)

d1 = distance(3, 10) # craeted object d1
d2 = distance(4, 4)  # created object d2
print("d1: {}  || d2: {}".format(d1, d2))

d3 = d1+d2 # invoking __add__ method where __add__(self, other)-->__add__(d1, d2)
print(d3)
```

```python
# @title __ ge __ method
# this is overload the ' >= ' operator

class distance:
  def __init__(self, x=None, y=None):
    self.ft = x
    self.inch = y

  def __ge__(self, other):
    val1 = self.ft * 12 + self.inch
    val2 = other.ft * 12 + other.inch
    if val1 >= val2:
      return "obj1 is greater than obj2"
    else:
      return "obj1 is not greater than obj2"

d1 = distance(3,10)
d2 = distance(7, 3)
print(d1 >= d2)
```
